```python
genius = lg.Genius('wtRGvsRU1ZVf8NFtC4kB0GlRfEt-UWQE0bNE9RGyIpy6cD_6PXq6PfuQ9mO7VgHZ', skip_non_songs=True, excluded_terms=["(Remix)", "(Live)"], remove_section_headers=True)
```

```python
def get_lyrics(arr, k):
    c = 0
    for name in arr:
        try:
            with open('C:/Users/J/Documents/Coding Projects/song_lyric_prediction/lyrics.txt','a') as file:  # Use file to refer to the file object
                songs = (genius.search_artist(name, max_songs=k, sort='popularity')).songs
                s = [song.lyrics for song in songs]
                file.write("\n<END>\n".join(s))
                c += 1
                print(f"Songs grabbed:{len(s)}")
        except:
            print(f"some exception at {name}: {c}")
```

```python
get_lyrics(["The Beatles","Cage The Elephant","Jack Johnson", "Taylor Swift", "Milky Chance", "Kanye West", "Pink Floyd", "Maroon 5", "One Republic", "Justin Bieber", "Jon Bellion"],15)
```

```python
lines = []
with open("C:/Users/J/Documents/Coding Projects/song_lyric_prediction/lyrics.txt","r") as fp:
    for line in fp:
        lines.append(line)
```

```python
while '\n' in lines:
    lines.remove('\n')
```

```python
from nltk import RegexpTokenizer
```

```python
tokens = []
for line in lines:
    line = line.lower()
    tokenizer = RegexpTokenizer('[a-z\']+')    
    tokens.extend(tokenizer.tokenize(line))
```
